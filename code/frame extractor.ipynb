```python
import cv2
import os
```

```python
# Read the video from specified path
cam = cv2.VideoCapture("C:\\Users\\BurnD\\Desktop\\yolov11\\dataset\\Throwing Videos\\BH\\BH2.mp4")

#Choose where to save the file
output_dir = "C:\\Users\\BurnD\\Desktop\\yolov11\\dataset\\Frames\\BH2"

#Create Folder if it doesn't exist
try:
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

except OSError as e:
    print('Error: Creating directory of data:', e)

# Starting Frame Number
currentframe = 1

if not cam.isOpened():
    print("Error: Could not open video file.")

while True:
    # Read frame-by-frame
    ret, frame = cam.read()
#CHANGE THE NAME OF THE SAVED FILE
    if ret:
        # Construct image filename and path
        filename = os.path.join(output_dir, f"[BH2]_frame{currentframe}.jpg")
        print(f'Creating... {filename}')

        # Save frame as image
        cv2.imwrite(filename, frame)

        # Increment frame counter
        currentframe += 1
    else:
        break

# Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```
